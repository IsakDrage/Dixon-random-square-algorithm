```python
def dixon_random_squares(n,B):
    Primes=list(primes(1,B+1))
    k=len(Primes)
    R=list(range(1,ceil(sqrt(n))))
    Z=[]
    A=[]
    while len(A)<k+1:
        z=randrange(1,ceil(sqrt(n)))^2%n
        z2=z
        ai=[]
        for p in Primes:
            aip=0
            while z%p==0:
                aip+=1
                z=z/p
            ai.append(aip)
        if z==1:
            Z.append(z2)
            A.append(ai)
    Am=matrix(GF(2),A)
    Ker=Am.kernel()
    Am=Am.change_ring(ZZ)
    for t in Ker:
        tg=matrix(t)
        tg=tg.change_ring(ZZ)
        x=1
        y=1
        for i in range(len(t)-1):
            y*=Primes[i]**(sum((tg)*(Am.column(i)))/2)
        for j in range(len(t)):
            x*=Z[j]**t[j]
        x2=x%n
        y2=y%n
        g=gcd(x2-y2,n)
        if g%n!=0 and g!=1:
            break
        else:
            pass
    return g
```
